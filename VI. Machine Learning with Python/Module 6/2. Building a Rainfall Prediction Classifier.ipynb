## Import required libraries

```python
from pathlib import Path

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from sklearn.model_selection import train_test_split, GridSearchCV, StratifiedKFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
```

```python
root_path = Path(".").absolute()
data_path = root_path / "data"
data_path.mkdir(exist_ok=True)

%matplotlib inline
```

## Load the data

```python
# url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/_0eYOqji3unP1tDNKWZMjg/weatherAUS-2.csv"

file = data_path / "weatherAUS_2.csv"
df = pd.read_csv(file)
df.head()
```

```python
df.count()
```

<p>Sunshine and cloud cover seem like important features, but they have a lot of missing values, far too many to impute their missing values.</p>

## Drop all rows with missing values

```python
df = df.dropna()
df.info()
```

<p>Since we still have 56k observations left after dropping missing values, we may not need to impute any missing values.</p>

```python
df.columns
```

## Data leakage considerations

<p>Consider the descriptions above for the columns in the dataset. Are there any practical limitations to being able to predict whether it will rain tomorrow given the available data?</p>
<p>Many columns describe today's weather, especially afternoon <code>3pm</code> readings, <code>Rainfall</code>, and <code>Raintoday</code>, but the target is tomorrow's rain. Using them would cause data leakage because that information would not be avaliable at prediction time.</p>

## Points to note 1

<p>Some of the features that would be inefficient in predicting tomorrow's rainfall is <code>Rainfall</code>, <code>RainToday</code>, and all <code>3pm</code> columns such as <code>WindDir3pm</code>, <code>WindSpeed3pm</code> and so on.</p>
<p>If we adjust our approach and aim to predict today's rainfall using historical weather data up to and including yesterday, that we can legitimately utilize all of the available features. This shift would be particularly useful for practical applications, such as deciding whether you will bike to work today.</p>
<p>With this new target, we should update the names of the rain columns accordingly to avoid confusion.</p>

```python
df = df.rename(columns={"RainToday": "RainYesterday", "RainTomorrow": "RainToday"})
```

## Data Granularity

<p>Would the weather patterns have the same predictability in vastly different locations in Australia? I would think not. The chance of rain in one location can be much higher than in another. Using all of the locations requires a more complex model as it needs to adapt to local weather patterns.</p>
<p>Let's see how many observations we have for each location, and see if we can reduce our attention to a smaller region.</p>

## Location selection

<p>You could do some research to group cities in the <code>Location</code> column by distance, which I've done for you behind the scenes. I found that Watsonia is only 15 km from Melbourne, and the Melbourne Airport is only 18 km from Melbourne.</p>
<p>Let's group these three locations together and use only their weather data to build our localized prediction model. Because there might still be some slight variations in the weather patterns we'll keep <code>Location</code> as a categorical variable.</p>

```python
df = df.loc[df.loc[:, "Location"].isin(["Melbourne", "MelbourneAirport", "Watsonia"])]
df.info()
```

<p>We still have 7557 records, which should be enough to build a reasonably good model. You could always gather more data if needed by partioning the data into similar locations or simply by updating it from the source to include a larger time frame.</p>

## Extracting a seasonality feature

<p>Now consider the <code>Date</code> column. We expect the weather patterns to be seasonal, having different predictability levels in winter and summer for example. There may be some variation with <code>Year</code> as well, but we'll leave that out for now.</p>
<p>Let's engineer a <code>Season</code> feature from <code>Date</code> and drop <code>Date</code> afterward, since it is most likely less informative than season. Instead of applying a Python function row by row, we'll use vectorized operations on the entire column for better performance.</p>

## Map the dates to seasons and drop the `Date` column

```python
# Use df["Date"] (not df.loc[:, "Date"]) so pandas replaces the column dtype with datetime64[ns].
# Use np.array (not a Python list) so we can index with a Series of season indices.
df["Date"] = pd.to_datetime(df["Date"])
df["Season"] = np.array(["Summer", "Autumn", "Winter", "Spring"])[(df["Date"].dt.month % 12) // 3]
df = df.drop(columns="Date")
df
```

<p>Looks like we have a good set of features to work with.</p>

## Define the feature and target dataframes

```python
X = df.drop(columns="RainToday", axis=1)
y = df.loc[:, "RainToday"]
```

## Display the counts of each class

```python
y.value_counts()
```

<p>The target variable is clearly imbalanced. There are 5766 "No" observations and 1791 "Yes" observations out of 7557 total records. This means it rains on about 23.7% of days in the Melbourne area, or roughly 86 to 87 rainy days per year. If we always predicted that it would not rain, we would still be correct about 76.3% (5766/7557) of the time, since "No" is the majority class. However, this does not mean the problem is easy to solve well for both classes. The dataset is imbalanced, with non-rainy days occurring about 3.2 times more often than rainy days.</p>
<p>For next steps, we should use a stratified train test split to preserve this class ratio, and evaluate the model using metrics beyond accuracy, such as precision, recall, F1 score, and the confusion matrix.</p>

## Split data into training and testing sets, ensuring target stratification

```python
split_data: list[pd.DataFrame | pd.Series] = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
X_train, X_test, y_train, y_test = split_data
```

## Define preprocessing transformers for numerical and categorical features

### Automatically detect numerical and categorical columns and assign them to separate numeric and categorical features

```python
numeric_features = X_train.select_dtypes(include=["number"]).columns.to_list()
categorical_features = X_train.select_dtypes(include=["object", "category"]).columns.to_list()
```

### Define separate transformers for both feature types and combine them into a single preprocessing transformer

```python
numeric_transformer = Pipeline(steps=[("scaler", StandardScaler())])
categorical_transformer = Pipeline(steps=[("onehot", OneHotEncoder(handle_unknown="ignore"))])
```

## Combine the transformers into a single preprocessing column transformer

```python
preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features)
])
```

## Create a pipeline by combining the preprocessing with a `RandomForestClassifier`

```python
pipeline = Pipeline(steps=[
    ("preprocessor", preprocessor),
    ("classifier", RandomForestClassifier(random_state=42))
])
```

### Define a parameter grid to use in a cross-validation grid search model optimizer

```python
param_grid = {
    "classifier__n_estimators": [50, 100],
    "classifier__max_depth": [None, 10, 20],
    "classifier__min_samples_split": [2, 5]
}
```

### Pipeline usage in cross-validation

<p>Recall that the pipeline is repeatedly used within the cross-validation by fitting on each internal training fold and predicting on its corresponding validation fold.</p>

## Perform grid search cross-validation and fit the best model to the training data

<p>Select a cross-validation method, ensuring target stratification during validation.</p>

```python
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
```

## Instantiate and fit `GridSearchCV` to the pipeline

```python
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, cv=cv, scoring="accuracy", verbose=2)
grid_search.fit(X_train, y_train)
```

### Print the best parameters and best cross-validation score

```python
print(f"Best parameters found: {grid_search.best_params_}")
print(f"Best cross-validation score: {grid_search.best_score_:.2f}")
```

## Display your model's estimated score

```python
print(f"Test set score: {grid_search.score(X_test, y_test):.2f}")
```

<p>So we have a reasonably accurate classifier, which is expected to correctly predict about 84% of the time whether it will rain today in the Melbourne area. But careful here. Let's take a deeper look at the results. The best model is stored within the grid search object.</p>

## Get the model predictions from the grid search estimator on the unseen data

```python
y_pred = grid_search.predict(X_test)
```

## Print the classification report

```python
print(f"Classification report: \n{classification_report(y_test, y_pred)}")
```

## Plot the confusion matrix

```python
conf_matrix = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix)
disp.plot(cmap="Blues")
plt.title("Confusion Matrix")
```

<p>Let's consider whether the results indicate a good predictor of rainfall.</p>

## Points to note 2

<p>The <b>True Positive Rate (TPR)</b>, also known as recall or sensitivity, measures how well the classifier identifies the positive class. It is defined as the ratio of true positives to all actual positive cases: </p>

$$
\text{TPR} = \frac{TP}{TP + FN}
$$

<p>In this rainfall prediction problem, the positive class is "Yes" (it will rain today).</p>
<p>Based on the classification report from the grid search model on the test set, the recall for the "Yes" class is 51%. Therefore, the true positive rate is 51%. Out of 358 test samples where it actually rained, the model correctly predicted rain for about 51% of them.</p>
<p>Although the overall test accuracy is 84%, this result shows that the model is much weaker at detecting rainy days than non-rainy days. The high accuracy is largely driven by the majority "No" class, while the model misses nearly half of the actual rain events. This is an important limitation to keep in mind when evaluating the classifier, and it is likely to appear in the assignment that follows.</p>

## Feature importances

<p>Recall that to obtain the categorical feature importances, we have to work our way backward through the modeling pipeline to associate the feature importances with their original input variables, not the one-hot encoded ones. We don't need to do this for the numeric variables because we didn't modify their names in any way.</p>
<p>Remember we went from categorical features to one-hot encoded features, using the <code>cat</code> column transformer.</p>

## Extract the feature importances

```python
feature_importances = grid_search.best_estimator_["classifier"].feature_importances_
```

<p>Now let's extract the feature importances and plot them as a bar graph.</p>

```python
feature_names = [*numeric_features, *grid_search.best_estimator_["preprocessor"].named_transformers_["cat"].named_steps["onehot"].get_feature_names_out(categorical_features).tolist()]

importance_df = pd.DataFrame({"Feature": feature_names, "Importance": feature_importances}).sort_values(by="Importance", ascending=False)

N = 20
top_features = importance_df.head(N)

plt.figure(figsize=(10, 6))
plt.barh(top_features.loc[:, "Feature"], top_features.loc[:, "Importance"], color="skyblue")
plt.gca().invert_yaxis()
plt.title(f"Top {N} Most Important Features in predicting whether it will rain today")
plt.xlabel("Importance Score")
```

## Point to note 3

<p><code>Humidity3pm</code> is the most important featrue for predicting whether it will rain today. Its importance score is clearly higher than that of the other features, with <code>Pressure3pm</code>, <code>Pressure9am</code> and <code>Sunshine</code> following behind.</p>

## Update the pipeline and the parameter grid

<p>Let's update the pipeline and the parameter grid and train a <code>LogisticRegression</code> model and compare the performance of the two models. You'll need to replace the classifier with <code>LogisticRegression</code>.</p>

```python
pipeline.set_params(**{"classifier": LogisticRegression(random_state=42)})
grid_search.estimator = pipeline

param_grid = {
    "classifier__solver": ["liblinear"],
    "classifier__penalty": ["l1", "l2"],
    "classifier__class_weight": [None, "balanced"]
}
grid_search.param_grid = param_grid

grid_search.fit(X_train, y_train)

y_pred = grid_search.predict(X_test)
```

## Compare the results to the previous model

```python
print(classification_report(y_test, y_pred))
```

```python
conf_matrix = confusion_matrix(y_test, y_pred)

plt.figure()
sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt="d")

plt.title("Titanic Classification Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")

plt.tight_layout()
```

## Points to note 4

<p>On the held-out test set of 1,512 samples, the <code>RandomForestClassifier</code> archived an overall accuracy of 84%, correctly predicting rainfall outcomes for approximately 1,270 of those cases. The <code>LogisticRegression</code> model performed slightly worse, with an accuracy of 83%, making roughly 1,250 correct predictions out of the same 1,512 test samples. Thus, Random Forest has a modest edge in overall accuracy, about 1% higher, and identifies about 20 more cases correctly than Logistic Regression.</p>
<p>Despite this difference in overall accuracy, both models share the same limitation when it comes to detecting actual rainy days. The <b>true positive rates (TPR)</b>, also known as recall for the positive "Yes" class is 51% for both classifiers. For Random Forest, this means that out of 358 test days when it actually rained, the model correctly predicted rain on only about 51% of them (roughly 183 cases). Logistic Regression shows the same 51% TPR for the "Yes" class, so it also misses nearly half of actual rain events. In other words, while Random Forest is marginally more accurate overall, neither model is substantially better at identifying rainy days. The high of both is driven largely by their strong performance on the majority "No" class (recall of 95% for Random Forest and 93% for Logistic Regression), rather than by improved detection of rain itself.</p>
