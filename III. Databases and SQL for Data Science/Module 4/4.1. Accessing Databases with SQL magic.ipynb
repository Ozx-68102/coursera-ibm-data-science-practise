# Access Databases with SQL Magic

<p>To communicate with SQL Databases from within a JupyterLab notebook, we can use the SQL "magic" provided by the <a href="https://github.com/catherinedevlin/ipython-sql">ipython-sql</a> extension. "Magic" is JupyterLab's term for special commands that start with a percentage sign <code>%</code>. Below, we'll use the <code>load_ext</code> magic to load the <code>ipython-sql</code> extension.</p>

<p>Here you will be creating and connecting to a new SQlite database <code>magic</code>.</p>

<p>The syntax for connecting to magic SQL using sqlite is:</p>
<p><b>%sql sqlite://DatabaseName</b></p>
<p>Where <code>DatabaseName</code> will be your <code>.db</code> file.</p>

```python
with sqlite3.connect(db_name) as con:
    con.cursor()
```

```python
%sql sqlite:///data/magic.db
```

## Using Python Variables in your SQL statements

<p>You can use python variables in your SQL statements by adding a ":" prefix to your python variable names. For example, if I have a python variable <code>country</code> with a value of <code>"Canada"</code>, I can use this variable in a SQL query to find all the rows of students from Canada.</p>

```python
country = "Canada"
%sql SELECT * FROM INTERNATIONAL_STUDENT_TEST_SCORES WHERE `COUNTRY`=:country;
```

## Assigning the Results of Queries to Python Variables

<p>You can use the normal python assignment syntax to assign the results of your queries to python variables. For example, I have a SQL query to retrieve the distribution of test scores (i.e. how many students got each score). I can assign the result of this query to the variable <code>test_score_distribution</code> using the <code>=</code> operator.</p>

```python
test_score_distribution = %sql SELECT `test_score` AS "Test_score", COUNT(*) AS "Frequency" FROM INTERNATIONAL_STUDENT_TEST_SCORES GROUP BY `test_score`;
test_score_distribution
```

## Converting Query Results to DataFrames

<p>You can easily convert a SQL query result to a pandas dataframe using the <code>DataFrame()</code> method. Dataframe objects are much more versatile than SQL query result objects. For example, we can easily graph our test score distribution after converting to a dataframe.</p>

```python
dataframe = test_score_distribution.DataFrame()

%matplotlib inline
plot = seaborn.barplot(x="Test_score", y="Frequency", data=dataframe)
```
